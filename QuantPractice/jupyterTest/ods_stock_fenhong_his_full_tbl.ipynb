```python
import pandas as pd
import numpy as np
```

```python
def func(money):
    if money == "--":
        return 0
    else:
        foo = money.replace('万', 'e4').replace('亿', 'e8')
        foo = eval(foo)
        return int(foo)
```

```python
df2["dividend"] = df2["dividend"].map(func)
```

```python
df2["dividend"] = df2["dividend"]/100000000
```

```python
df2.head()
```

```python
sum(list(df2['dividend']))
```
